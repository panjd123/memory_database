```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('SSB_benchmark_merged_old.csv')
df["TestCase"] = df["TestCase"].astype(str)  # 确保 TestCase 列是字符串类型
df["TestCase"] = pd.Categorical(df["TestCase"],categories=["3.1", "3.2", "3.3", "4.1", "4.2", "2.1", "2.2", "2.3", "4.3", "3.4"], ordered=True)

# df = df[df["Method"] == "columnwise_dynamic_vector"]
df = df[df["Method"] == "rowwise"]
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=df, x="TestCase", y="Time", hue="code", style="code", markers=True, linewidth=2, markersize=8, ax=ax)
for line in ax.get_lines():
    x_data = line.get_xdata()
    y_data = line.get_ydata()
    for x, y in zip(x_data, y_data):
        label = f"{y:.3f}"
        ax.text(x, y, label, fontsize=8, ha='left', va='bottom')
```
